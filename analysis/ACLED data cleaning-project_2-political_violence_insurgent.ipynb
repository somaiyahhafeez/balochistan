```python
import pandas as pd  
```

## Reading CSV file. 

```python
df = pd.read_csv("asia_pacific_acled.csv")
df.head()
```

## Filtering for Balochistan and selecting events after 2020. Starting with events of political violence.

```python
df_balochistan = df[df["admin1"] == "Balochistan"]
df_balochistan_2020 = df_balochistan[df_balochistan["event_date"] >= "2022-01-01"]
df_balochistan_2020_political_violence = df_balochistan_2020[df_balochistan_2020["disorder_type"] == "Political violence"]
```

```python
df_balochistan_2020_political_violence['actor1'].unique()
```

```python
df_balochistan_2020_political_violence['inter1'].unique()
```

## grouping by months.

```python
df_balochistan_2020_political_violence["event_date"] = pd.to_datetime(
    df_balochistan_2020_political_violence["event_date"], errors="coerce"
)
df_balochistan_2020_political_violence["month"] = df_balochistan_2020_political_violence["event_date"].dt.to_period("M")
```

```python
df_balochistan_2020_political_violence
```

```python
df_balochistan_2020_political_violence=df_balochistan_2020_political_violence.groupby(["inter1", "month"]).size().reset_index(name="event_count")
df_balochistan_2020_political_violence
```

## filtering for baloch separatists groups

```python
df_filtered = df_balochistan_2020_political_violence[df_balochistan_2020_political_violence['inter1'].isin(['Rebel group'])]

df_filtered = df_filtered[df_filtered['month'] != '2025-02-01']
df_filtered
```

```python
df_filtered.to_csv("political_violence_balochistan_insurgents.csv")
```
