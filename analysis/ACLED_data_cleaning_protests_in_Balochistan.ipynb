```python
import pandas as pd  
```

## Reading file

```python
df = pd.read_csv("asia_pacific_acled.csv")
df.head()
```

## Filtering for BYC protests in Pakistan

```python
df = df[df["country"] == "Pakistan"]
```

```python
df = df[(df["event_date"] >= "2023-11-01") & (df["event_date"] <= "2024-02-29")]
```

```python
df_demonstrations = df[
    df["disorder_type"].isin(["Political violence; Demonstrations", "Demonstrations"])
]
```

```python
df_byc = df[df["assoc_actor_1"].str.contains("BYC", na=False, case=False)]
df_byc
```

## Keeping some columns only.

```python
columns_to_keep=["event_date", "location", "latitude", "longitude"]
```

```python
df_byc = df_byc[columns_to_keep]
```
